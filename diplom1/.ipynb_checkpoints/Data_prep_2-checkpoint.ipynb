Судя по графикам распределений, числовые признаки необходимо будет нормализовть и логaрифмировать для получения нормального распределения

Проверим на мультиколлинеарность и корреляцию числовых признаков между собой и признаком 'target'

```python
corr_matrix = np.corrcoef(data[numfeatures], rowvar=False)
print('Ранг матрицы корреляций:', np.linalg.matrix_rank(corr_matrix))
print('Определитель матрицы корреляций: {:.7f}'.format(np.linalg.det(corr_matrix)))
sns.heatmap(corr_matrix, xticklabels=numfeatures, yticklabels=numfeatures , annot=True, cmap='coolwarm')
```

Перейдем к категориальны и биноминальным признакам.
Для работы с ними я буду использовать для наглядности данные, из которых исключены строки в которых целевой признак
более 1000000. При этом удалять из датасета их считаю не правильным, т.к. выбросами они не являются(площадь, количество ванных, местоположение соответствует премиальному жилью).

```python
data1 = data[data['target'] < 1000000]
```

```python
n =  len(catfeatures)
fig, axes = plt.subplots(n, 1, figsize=(15, 50)) # фигура + n х 1 координатных плоскостей
for i, feature in enumerate(catfeatures): #i- индекс, feature - название признака в таблице

    barplot = sns.boxplot(x=feature, y='target',data=data1.loc[data1.loc[:, feature].isin(data1.loc[:, feature].value_counts().index[:15])], ax=axes[i])
    
    #Добавляем подпись графику
    barplot.set_title('{} VS target'.format(feature));
    #Добавляем поворот отметок на оси абсцисс
    barplot.xaxis.set_tick_params(rotation=45)
    
    
```

```python
n =  len(binfeatures)
fig, axes = plt.subplots(n, 1, figsize=(15, 50)) # фигура + n х 1 координатных плоскостей
for i, feature in enumerate(binfeatures): #i- индекс, feature - название признака в таблице

    barplot = sns.boxplot(x=feature, y='target',data=data1.loc[data1.loc[:, feature].isin(data1.loc[:, feature].value_counts().index[:15])], ax=axes[i])
    
    #Добавляем подпись графику
    barplot.set_title('{} VS target'.format(feature));
    #Добавляем поворот отметок на оси абсцисс
    barplot.xaxis.set_tick_params(rotation=45)
```

Проведем статистический анализ значимости для категориальных и биноминальных признаков.

```python
for column in catfeatures:
    pv = []
    cols = data.loc[:, column].value_counts().index[:15]
    comb_all = list(combinations(cols, 2))
    for comb in comb_all:
        pv.append(ttest_ind(data[data[column] == comb[0]]['target'], data[data[column] == comb[1]]['target']).pvalue)
        

    if min(pv)<= 0.05/len(comb_all):
        print('Колонка ', column, ' статистически значима')
```

```python
for column in binfeatures:
    pv = []
    cols = data.loc[:, column].value_counts().index
    comb_all = list(combinations(cols, 2))
    for comb in comb_all:
        pv.append(ttest_ind(data[data[column] == comb[0]]['target'], data[data[column] == comb[1]]['target']).pvalue)
        

    if min(pv)<= 0.05/len(comb_all):
        print('Колонка ', column, ' статистически значима')
```

Как можно видеть все признаки являются значимыми.
